```python
# pynpl
from pynlpl.textprocessors import crude_tokenizer, Windower, swap, Tokenizer
from pynlpl.statistics import FrequencyList

# Nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer


```

# Processamento de texto

```python
trecho_cronica = """
A cartomante fê-lo sentar diante da mesa, e sentou-se do lado oposto, com as costas para a janela, de maneira que a pouca luz de fora batia em cheio no rosto de Camilo.
Abriu uma gaveta e tirou um baralho de cartas compridas e enxovalhadas. Enquanto as baralhava, rapidamente, olhava para ele, não de rosto, mas por baixo dos olhos.
ra uma mulher de quarenta anos, italiana, morena e magra, com grandes olhos sonsos e agudos. Voltou três cartas sobre a mesa, e disse-lhe:
— Vejamos primeiro o que é que o traz aqui. O senhor tem um grande susto…
"""

tokens = crude_tokenizer(trecho_cronica)
print(tokens)
```

# Anagramas

N-grams podem ser usados como características em modelos de aprendizado de máquina para tarefas de classificação de texto, análise de sentimento, detecção de spam, entre outros. Cada n-gram pode ser tratado como uma característica independente.

```python
for trigram in Windower(tokens,3, None, None):
        print (" ".join(trigram))
```

```python
freqlist = pynlpl.statistics.FrequencyList()
freqlist.append(tokens)

freqlist_ordenada = sorted(freqlist, key=lambda x: x[1], reverse=True)

top5_palavras = freqlist_ordenada[:15]

for palavra, contagem in top5_palavras:
    print(f"{palavra}: {contagem}")


```

## Significado das tags



1. PROPN (Proper Noun): Nome próprio, como nomes de pessoas, lugares, organizações, etc. Exemplos: "João", "Brasil", "Google".

2. AUX (Auxiliary Verb): Verbo auxiliar, usado em construções gramaticais para indicar tempo, modo, aspecto, etc. Exemplos: "estar", "ser", "ter".

3. VERB (Verb): Verbo principal, que expressa uma ação, estado ou ocorrência. Exemplos: "correr", "falar", "comer".

4. N,OUN (Noun): Substantivo, que é uma palavra usada para nomear uma pessoa, lugar, coisa ou ideia. Exemplos: "casa", "gato", "felicidade".

5. ADV (Adverb): Advérbio, que modifica ou qualifica um verbo, adjetivo ou outro advérbio. Exemplos: "rapidamente", "bem", "muito".

6. ADP (Adposition): Adposição, que é uma palavra ou prefixo que expressa uma relação sintática ou semântica entre uma palavra e uma palavra subsequente na frase. Exemplos: "em", "para", "sobre".

7. NUM (Numeral): Numeral, que é uma palavra que denota um número ou uma quantidade. Exemplos: "um", "dois", "três".

8. SYM (Symbol): Símbolo, que inclui caracteres, sinais de pontuação e outros símbolos especiais. Exemplos: "$", "#", "@".
